```julia
using DataFrames
using Plots
```

```julia
df = readtable("results.csv", header = false, names=[:Place, :DivTotal, :Div, :Guntime, :Nettime, :Pace])
```

```julia
pace = df[:][6]
```

```julia
function ConverPaceToSpeed(pace)
    
    minutes = zeros(length(pace), 1)
    seconds = zeros(length(pace), 1)
    
    ## splitting pace column into minutes and seconds
    for i = 1:length(pace)
        if(length(pace[i][:]) == 5)  # width of the row in a pace vector
            minutes[i] = parse(Int, pace[i][1:end-3]) 
            seconds[i] = parse(Int, pace[i][4:5])
        else 
            minutes[i] = parse(Int, pace[i][1:end-3])
            seconds[i] = parse(Int, pace[i][3:4])
          end
    end
    
    seconds = minutes*60 + seconds
    speed = [1/i*60*60 for i in seconds]
    
    return speed
end
```

```julia
speed = ConverPaceToSpeed(pace)
```

```julia
c = countmap(speed) 
```

```julia
## calc cdf
function calc_cdf(c) # c is a data dict
    runsum = 0
    xs = Float64[]
    cs = Float64[]

    for (value, count) in sort(c)
        runsum = runsum + count
        push!(xs, value)
        push!(cs, runsum)
    end

    total = float(runsum)
    cdf = [c/total for c in cs]   ## cdf[end] = 1, ps carrried cdf
    
    return xs, cdf
end
```

```julia
xs, cdf = calc_cdf(c)
```

```julia
plot(xs, cdf)
xlabel!("Values")
ylabel!("CDFs")
```

```julia
## let's calc the percentile rank .. 

function Percentile_Rank(scores, your_score)
    count = 0
    
    for score in scores
        if score <= your_score
            count += 1
        end
    end

    percentile_rank = 100.0 * count / length(scores)
    
    return percentile_rank
end
```

```julia
author_speed = speed[end] 
```

```julia
Percentile_Rank(speed, author_speed)
```

Matches Python results .. 

and it is a reasonable result .. as the rank of the author in the race (the last) .. 

hence, all the competitors were better then him .. 

this is why his percentile rank was very tiny .. 

To assure this way of thinking .. let's calc the percentile rank of the winner (speeds[0]) ..  that is expected to be 100% .. which means that he beats 100% of the participants .. 

```julia
perc_rank = Percentile_Rank(speed, speed[1])
perc_rank
```

As expected .. !
