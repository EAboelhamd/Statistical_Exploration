## Data Extraction:

```julia
df = readtable("results.csv", header = false, names=[:Place, :DivTotal, :Div, :Guntime, :Nettime, :Pace])
```

```julia
pace = df[:][6]
```

```julia
parse(Int, pace[2][3:4])
pace[1596][1:end-3]
```

```julia
function ConverPaceToSpeed(pace)
    
    minutes = zeros(length(pace), 1)
    seconds = zeros(length(pace), 1)
    
    ## splitting pace column into minutes and seconds
    for i = 1:length(pace)
        if(length(pace[i][:]) == 5)  # width of the row in a pace vector
            minutes[i] = parse(Int, pace[i][1:end-3]) 
            seconds[i] = parse(Int, pace[i][4:5])
        else 
            minutes[i] = parse(Int, pace[i][1:end-3])
            seconds[i] = parse(Int, pace[i][3:4])
          end
    end
    
    seconds = minutes*60 + seconds
    speed = [1/i*60*60 for i in seconds]
    
    return speed
end
```

```julia
speed = ConverPaceToSpeed(pace)
```

```julia
c = countmap(speed) 
```

```julia
length(speed)
```

```julia
## calc cdf
function calc_cdf(c) # c is a data dict
    runsum = 0
    xs = Float64[]
    cs = Float64[]
   
    for (value, count) in sort(c)
        runsum = runsum + count
        push!(xs, value)
        push!(cs, runsum)
    end

    total = float(runsum)  # total = 1596
    cdf = [c/total for c in cs]   ## cdf[end] = 1, ps carrried cdf
    
    return xs, cdf
end
```

```julia
xs, cdf = calc_cdf(c)
```

```julia
plot(xs, cdf)
xlabel!("Values")
ylabel!("Probabilties")
```

```julia
function calc_pmf(c)
    runsum_ = 0
    keys = Float64[]
    vals = Float64[]

    for (k, v) in c
        runsum_ = runsum_ + v
        push!(keys, k) 
        push!(vals, v) 
    end
    
    
    total = float(runsum_)  # total = 1596
    pmf = [p/total for p in vals] 
    
    return keys, pmf
end
```

```julia
index, pmf = calc_pmf(c)
```

```julia
bar(index, pmf)
xlabel!("Speeds")
ylabel!("PMF")
```

As demonestrated in python plot .. the distrubution of the speeds has a normal shape. 

It is clear that PMF plot gives a better demonestration for the shape of the distribution ..
