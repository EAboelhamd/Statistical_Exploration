## Calc. PMF for 1st babies and others:

```julia
c = countmap(first_)  ## Data dictionary for first babies ..
```

```julia
function ProbEarly(index, pmf)
    k_ = []
    v_ = []
    
    for i = 1:length(index)
        if(index[i] <= 37)
            append!(k_, index[i])
            append!(v_, pmf[i])
        end
    end
    return k_, v_
end
```

```julia
function ProbOnTime(index, pmf)
    k__ = []
    v__ = []
    
    for i = 1:length(index)
        if(index[i] <= 40 && index[i] >= 38)
            append!(k__, index[i])
            append!(v__, pmf[i])
        end
    end
    return k__, v__
end
```

```julia
function ProbLate(index, pmf)
    k___ = []
    v___ = []
    
    for i = 1:length(index)
        if(index[i] >= 41)
            append!(k___, index[i])
            append!(v___, pmf[i])
        end
    end
    return k___, v___
end
```

```julia
pmf = []
index = []

for (k, v) in c
    append!(pmf, v/length(first_))
    append!(index, k)       
end
```

```julia
k_, v_ = ProbEarly(index, pmf)
```

```julia
bar!(k_, v_)
```

## Comment:
The above output vector for PMF does make sense as its indices within the range ..  

```julia
k__, v__ = ProbOnTime(index, pmf)
```

```julia
bar!(k__, v__)
```

```julia
k___, v___ = ProbLate(index, pmf)
```

```julia
bar!(k___, v___)
```

## Computing fraction of births in each range .. 

```julia
fraction_early = sum(v_)
fraction_on_time = sum(v__)
fraction_late = sum(v___)
fraction_early, fraction_on_time, fraction_late
```

## Comment:

The above fractions indicate the majority of first_babies come on time .. and it is rare to have an early first_baby ..
