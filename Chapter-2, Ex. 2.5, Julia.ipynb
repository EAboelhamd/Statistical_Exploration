```julia
using DataFrames
```

We gonna copy the same list of elements as Python code .. just to be able to compare the results .. 

```julia
data = [30, 20, 25, 40, 50, 25]
data
```

```julia
c = countmap(data)  ## create data dictionary 
```

## 1. Calc Mean for PMF:

```julia
function PmfMean(c)
    pmf = []
    index = []

    for (k, v) in c
        append!(pmf, v/length(data))
        append!(index, k)       
    end
    
    return sum(index.*pmf)
end
```

```julia
m = PmfMean(c)
```

- The mean value for PMF generated by Julia matches that one generated in Python as well as the built in function implemented by the uthor of the textbook .. 

- I think the value of 31.667 makes sense .. as the distribution of the values is nearly normal .. and we have not got any outliers hence, it is expected to have a mean value close to the majority of the data points .. 

## 1. Calc Variance for PMF:

```julia
function PmfVar(c)
    pmf = []
    index = []

    for (k, v) in c
        append!(pmf, v/length(data))
        append!(index, k)       
    end
    
    return sum(pmf.*(index - m).^2)
end
```

```julia
PmfVar(c)
```
