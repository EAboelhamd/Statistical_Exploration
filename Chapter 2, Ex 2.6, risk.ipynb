```python
all_ages_prob = Pmf.MakePmfFromList(all_ages)
all_ages_prob
```

```python
# ProbEarly (PMF_) ... return age <= 37

def ProbEarly (all_ages_prob):
    
    PMF_ = all_ages_prob.d.values()
    ages = all_ages_prob.d.keys()
    early_v = []
    early_k = []
    
    for j in xrange(len(ages)):
        if(ages[j] <= 37):
            early_v.append(PMF_[j])
            early_k.append(ages[j])
    return early_v, early_k
```

```python
ProbEarly(all_ages_prob)
```
